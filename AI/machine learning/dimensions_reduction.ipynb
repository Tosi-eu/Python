```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('364.jpg')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

soft_image = cv2.GaussianBlur(gray_image, (7, 7), 0)

(T, binary_image) = cv2.threshold(soft_image, 160, 255, cv2.THRESH_BINARY)
(T, binary_inv_image) = cv2.threshold(soft_image, 160, 255, cv2.THRESH_BINARY_INV)

titles = ['Imagem em Tons de Cinza', 'Imagem Binarizada', 'Imagem Binarizada Invertida']
images = [soft_image, binary_image, binary_inv_image]

plt.figure(figsize=(12, 4))

for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
