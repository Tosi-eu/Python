```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

VP = 50
FN = 10
FP = 5
VN = 100
N = VP + FN + FP + VN

conf_matrix = np.array([[VP, FN], [FP, VN]])
```

```python
def print_conf_matrix(matrix):
  plt.figure(figsize=(8, 6))
  sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Spam', 'Não Spam'], yticklabels=['Spam', 'Não Spam'])
  plt.xlabel('Previsto')
  plt.ylabel('Real')
  plt.title('Matriz de Confusão')
  plt.show()

def get_sensibility(true_positive, false_negative):
  return true_positive / (true_positive + false_negative)

def get_especificity(true_negative, false_positive):
  return true_negative / (true_negative + false_positive)

def get_accuracy(true_positive, true_negative, total_elements):
  return (true_positive + true_negative) / total_elements

def get_precision(true_positive, true_negative):
  return true_positive / (true_positive / true_negative)
```

```python

sensibilidade = get_sensibility(VP, FN)
especificidade = get_especificity(VN, FN)
acuracia = get_accuracy(VP, VN, N)
precisao = get_precision(VP, VN)

# F-Score
f_score = (2 * precisao * sensibilidade) / (precisao + sensibilidade)

# Imprimindo os resultados
print(f'Sensibilidade: {sensibilidade:.2f}')
print(f'Especificidade: {especificidade:.2f}')
print(f'Acurácia: {acuracia:.2f}')
print(f'Precisão: {precisao:.2f}')
print(f'F-Score: {f_score:.2f}')
```

```python
print_conf_matrix(conf_matrix)
```
